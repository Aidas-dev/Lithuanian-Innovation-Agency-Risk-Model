### Creating total employees column and a single average wage column.

```python
# employees
create_variable_eval(E, 'employees', '`numInsured` + `numInsured2`', fill_na=0)


# wages
create_variable_eval(E, 'average_wage', '(`avgWage` * `numInsured` + `avgWage2` * `numInsured2` ) / (`numInsured` + `numInsured2`)', fill_na=0)


# removing redundant columns
remove_columns(E, ['numInsured', 'numInsured2', 'avgWage', 'avgWage2'], inplace=True)

set_dataframe_values(E, 0, np.nan, columns='average_wage', inplace=True)

# Rounding everything to 2 decimals:

round_dataframe_values(E, 2, inplace=True)
```

### Pivoting the data so that each row represents a unique firm-month combination with employment numbers as values.

```python
# Extracing municipality, ja_kodas and ecoActCode for later merging back:

EforMerge = extract_columns(E, ['ja_kodas', 'municipality', 'ecoActCode', 'ecoActName'])

# Removing the columns from the main dataframes:
remove_columns(E, ['municipality', 'ecoActCode', 'ecoActName'])
```

```python

# Pivoting the data:

Epivot = pivot_dfs_smart_flexible(
    dataframes=E,
    pivot_column='month',
    value_columns=['average_wage', 'employees'],  # Multiple value columns as list
    id_columns='ja_kodas',                        # Single ID column as string
    verbose=True,
    inplace=False
)
```

### Calculating the yearly averages for employment and wages.

```python
# Most straightforward approach - process each DataFrame directly
for i, df in enumerate(Epivot):
    print(f"Processing DataFrame {i}...")

    # Get ALL columns with the prefixes (will work for any year)
    wage_cols = [col for col in df.columns if 'average_wage_' in col]
    emp_cols = [col for col in df.columns if 'employees_' in col]

    # Use simple pandas mean (most reliable)
    if wage_cols:
        df['yearly_average_wage'] = df[wage_cols].mean(axis=1, skipna=True)

    if emp_cols:
        df['yearly_average_employees'] = df[emp_cols].mean(axis=1, skipna=True)

    print(f"  Added averages using {len(wage_cols)} wage columns and {len(emp_cols)} employee columns")

print("All DataFrames processed!")
print(Epivot[0][['ja_kodas', 'yearly_average_wage', 'yearly_average_employees']].head())
```

```python
# Extracting the averages for later merging back:
EyearlyAvg = extract_columns(Epivot, ['ja_kodas', 'yearly_average_wage', 'yearly_average_employees'])
```
